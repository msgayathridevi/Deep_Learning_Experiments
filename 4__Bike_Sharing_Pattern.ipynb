```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("train.csv")
df.head()
```

CONVERSION OF STRING DATA SET

```python
df["datetime"] = df["datetime"].astype("datetime64")
```

```python
df["hour"] = df["datetime"].dt.hour
df["day"] = df["datetime"].dt.day
df["month"] = df["datetime"].dt.month
df["year"] = df["datetime"].dt.year
```

```python
df.head()
```

```python
df = df.drop(["datetime"], axis=1)
```

```python
y = df["count"]
X = df.drop(["count"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=64)
```

NORMALIZING THE DATA SET

```python
data = (X_train - X_train.mean())/X_train.std()
y.head()
```

```python
X_train
```

A Sequential model is appropriate for a plain stack of layers where each layer has exactly one input tensor and one output tensor.

```python
model = Sequential()
# model.add(Dense(128, activation="relu"))
# model.add(Dense(20, activation="linear"))
# model.add(BatchNormalization())
model.add(Dense(16, activation="linear"))
model.add(Dense(1))
model.compile(loss="mse", optimizer="adam")  #gradient descent
```

```python
model.fit(X_train,y_train, validation_split=0.3, epochs = 60, verbose = 1)
#the choice that how you want to see the output of your Nural Network while it's training.
```

```python
model.evaluate(X,y)
```
